### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python


# Calcular los ingresos por tienda
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()



```

```python
# Mostrar resultados
print(" Ingreso total por tienda:")
print(f"Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda4:,.2f}")
```

```python
# Calcular el ingreso total global (opcional)
total_general = ingreso_tienda1 + ingreso_tienda2 + ingreso_tienda3 + ingreso_tienda4
print(f"\n Ingreso total de todas las tiendas: ${total_general:,.2f}")
```

# 2. Ventas por categoría

```python
# Calcular cantidad de productos vendidos por categoría (cuenta de filas por categoría)
ventas_t1 = tienda.groupby("Categoría del Producto")["Precio"].count().sort_values(ascending=False)
ventas_t2 = tienda2.groupby("Categoría del Producto")["Precio"].count().sort_values(ascending=False)
ventas_t3 = tienda3.groupby("Categoría del Producto")["Precio"].count().sort_values(ascending=False)
ventas_t4 = tienda4.groupby("Categoría del Producto")["Precio"].count().sort_values(ascending=False)
```

```python
# Mostrar los resultados
print(" Ventas por categoría - Tienda 1:")
print(ventas_t1, "\n")

print(" Ventas por categoría - Tienda 2:")
print(ventas_t2, "\n")

print(" Ventas por categoría - Tienda 3:")
print(ventas_t3, "\n")

print("Ventas por categoría - Tienda 4:")
print(ventas_t4, "\n")
```

# 3. Calificación promedio de la tienda


```python
# Calcular el promedio de calificación por tienda
prom_tienda1 = tienda["Calificación"].mean()
prom_tienda2 = tienda2["Calificación"].mean()
prom_tienda3 = tienda3["Calificación"].mean()
prom_tienda4 = tienda4["Calificación"].mean()
```

```python
# Mostrar los resultados con formato
print(" Calificación promedio por tienda:")
print(f"Tienda 1: {prom_tienda1:.2f}")
print(f"Tienda 2: {prom_tienda2:.2f}")
print(f"Tienda 3: {prom_tienda3:.2f}")
print(f"Tienda 4: {prom_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Función para mostrar top y bottom productos por tienda
def analizar_productos(df, nombre_tienda):
    conteo = df.groupby("Producto")["Precio"].count().sort_values(ascending=False)

    print(f" {nombre_tienda}")
    print("Productos más vendidos:")
    print(conteo.head(5))
    print("\n Productos menos vendidos:")
    print(conteo.tail(5))
    print("-" * 50)

    #  Visualización (Top 5 más vendidos)
    conteo.head(5).plot(kind='bar', figsize=(7,4), title=f"Top 5 productos más vendidos - {nombre_tienda}", color='skyblue')
    plt.ylabel("Cantidad de ventas")
    plt.xlabel("Producto")
    plt.show()
```

```python
# Ejecutar análisis para cada tienda
analizar_productos(tienda, "Tienda 1")
analizar_productos(tienda2, "Tienda 2")
analizar_productos(tienda3, "Tienda 3")
analizar_productos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio para cada tienda
envio_prom_t1 = tienda["Costo de envío"].mean()
envio_prom_t2 = tienda2["Costo de envío"].mean()
envio_prom_t3 = tienda3["Costo de envío"].mean()
envio_prom_t4 = tienda4["Costo de envío"].mean()
```

```python
# Mostrar los resultados
print(" Costo de envío promedio por tienda:")
print(f"Tienda 1: ${envio_prom_t1:.2f}")
print(f"Tienda 2: ${envio_prom_t2:.2f}")
print(f"Tienda 3: ${envio_prom_t3:.2f}")
print(f"Tienda 4: ${envio_prom_t4:.2f}")
```

```python
# ---GRÁFICO DE BARRAS: Ingresos totales por tienda ---
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

plt.figure(figsize=(7,5))
plt.bar(ingresos.keys(), ingresos.values(), color=["skyblue", "lightgreen", "salmon", "orange"])
plt.title(" Ingreso total por tienda")
plt.ylabel("Monto total en $")
plt.xlabel("Tienda")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# ---  GRÁFICO DE PIE (CIRCULAR): Distribución de categorías en Tienda 1 ---
categorias_t1 = tienda["Categoría del Producto"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(categorias_t1, labels=categorias_t1.index, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title(" Distribución de categorías de productos - Tienda 1")
plt.show()
```

```python
# --- GRÁFICO DE LÍNEAS: Calificación promedio y costo de envío ---
calificaciones = [
    tienda["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

costos_envio = [
    tienda["Costo de envío"].mean(),
    tienda2["Costo de envío"].mean(),
    tienda3["Costo de envío"].mean(),
    tienda4["Costo de envío"].mean()
]

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(8,5))
plt.plot(tiendas, calificaciones, marker='o', label=' Calificación promedio', color='blue')
plt.plot(tiendas, costos_envio, marker='s', label=' Costo de envío promedio', color='red')
plt.title("Comparación entre calificaciones y costo de envío")
plt.xlabel("Tiendas")
plt.ylabel("Valor promedio")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import pandas as pd

# URLs de las tiendas
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar datos
t1 = pd.read_csv(url1)
t2 = pd.read_csv(url2)
t3 = pd.read_csv(url3)
t4 = pd.read_csv(url4)

# Calcular ingresos por tienda
ingresos = {
    "Tienda 1": t1["Precio"].sum(),
    "Tienda 2": t2["Precio"].sum(),
    "Tienda 3": t3["Precio"].sum(),
    "Tienda 4": t4["Precio"].sum()
}

# Mostrar resultados ordenados
ingresos_df = pd.DataFrame(list(ingresos.items()), columns=["Tienda", "Ingreso Total"]).sort_values("Ingreso Total")
print(" Ingreso total por tienda:")
print(ingresos_df)

# Identificar la tienda con menos lucro
tienda_menor = ingresos_df.iloc[0]
print(f"\n La tienda que menos lucro genera es {tienda_menor['Tienda']} con un ingreso total de {tienda_menor['Ingreso Total']:.2f}")
```
